```python
IMAGE_SHAPE = (224, 224)
```

```python
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import HTML, display ,Image
```

```python
image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255)
training_data='/content/drive/MyDrive/AI base Alzheimer s care and Cognitive Support Mobile App  /Randani DOC/train'
training_image_data  = image_generator.flow_from_directory(training_data,target_size=IMAGE_SHAPE)
```

```python
class_names = sorted(training_image_data.class_indices.items(), key=lambda pair:pair[1])
class_names = np.array([key.title() for key, value in class_names])
class_names
```

```python
from ultralytics import YOLO

model = YOLO('yolo11n-cls.pt')

results = model.train(data='', epochs=20, imgsz=640)
```

```python
classes=class_names
```

```python
model = YOLO('runs/classify/train3/weights/best.pt')

results = model("/content/drive/MyDrive/AI base Alzheimer s care and Cognitive Support Mobile App  /Randani DOC/train/incorrect colock 10 ,11 numeral marking/WhatsApp Image 2024-11-21 at 12.46.39 AM.jpeg")

print(results)
```

```python
def predict_image(url):
  results = model(url)
  if results[0].probs.top5conf[0]>0.5:
    top_prediction = results[0].probs.top1
    text = f"Predicted: {str(classes[top_prediction]).title()}"
    display(HTML(f"<h2 style='color: green;'>Prediction Result:</h2>"))
    display(HTML(f"<p style='color: red; font-size: 24px;'>{text}</p>"))
    display(Image(filename=results[0].path, width=400, height=300))
  else:
    display(HTML(f"<h2 style='color: red;'>Prediction Result:Not Detected</h2>"))
    display(Image(filename=results[0].path, width=400, height=300))
```

```python
predict_image('/content/drive/MyDrive/AI base Alzheimer s care and Cognitive Support Mobile App  /Randani DOC/train/incorrect pyramid-templet/WhatsApp Image 2024-11-21 at 6.14.49 PM.jpeg')
```
